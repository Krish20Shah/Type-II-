```python
#!/home/krish.shah/anaconda3/envs/Teobresums/bin/python3

import numpy as np
import matplotlib.pyplot as plt

```

```python
GG = np.loadtxt('FF_value_HM.txt')
q_inj = np.linspace(2,10,20) #mass ratio values 
l_inj = np.linspace(np.pi/6,np.pi/2,20)
coordinates = []
for i in range(len(q_inj)):
    for j in range(len(l_inj)):
        coordinates.append((q_inj[i], l_inj[j]))
coordinates = np.array(coordinates)

plt.tricontourf(coordinates[:, 0], coordinates[:, 1], GG.flatten(),1000,alpha = 1,cmap = 'plasma_r')
cbar = plt.colorbar()
plt.xlabel('Mass ratio')
plt.ylabel('Incilnation')
cbar.set_label('FF_Value')
plt.show()
```
